```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=np.arange(100,850,50)
T
Cp=[40.144,41.826,43.396,44.860,46.222,47.489,48.665,49.755,50.764,51.698,52.559,53.353,54.084,54.755,55.371]
```

```python
T=T+273.15 #Temperaturas de Â°C a K
y=Cp
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
print(beta)
Cp_cal=X.dot(beta)
Cp_cal #Cp calculados
```

```python
plt.scatter(T,y,c='black')
plt.plot(T,Cp_cal)
plt.show()
```

```python
dT=0.1
T=np.arange(100,800,dT)
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
Cp_cal=X.dot(beta)
plt.plot(T,Cp_cal)
plt.show()
```

```python
H=np.cumsum(Cp_cal*dT)
plt.plot(T,H,color='purple')
plt.xlabel('Temperatura (K)')
plt.ylabel('delta_H (kJ/mol)')
plt.title('delta_H vs Temperatura')
plt.show()
```

```python
S=np.cumsum(Cp_cal/T*dT)
plt.plot(T,S,color='')
plt.xlabel('Temperatura (K)')
plt.ylabel('delta_S (J/K-mol)')
plt.title('delta_S vs Temperatura')
plt.show()
```

```python
G=np.cumsum(H-T*S)
plt.plot(T,G,color='purple')
plt.xlabel('Temperatura (K)')
plt.ylabel('delta_G (kJ/mol)')
plt.title('delta_G vs Temperatura')
plt.show()
```
