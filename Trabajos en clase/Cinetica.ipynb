```python
import matplotlib.pyplot as plt
import numpy as np
```

## CinÃ©tica de primer orden
## A --> B

```python
A=1.0
B=0
k1=0.1
dt=0.1
t=0
x=[]
yA=[]
yB=[]
while t<100:
    x.append(t)
    yA.append(A)
    yB.append(B)
    dA=-k1*A*dt
    dB=+k1*A*dt
    A+=dA
    B+=dB
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel('Tiempo(min)')
plt.ylabel('ConcentaciÃ³n')
plt.title('Cinetica de primer orden')
plt.show()
```

## CinÃ©tica secuencial
## A--> B
## B--> C

```python
A=1.0
B=0
C=0
k1=0.1
k2=0.11
dt=0.1
t=0
x=[]
yA=[]
yB=[]
yC=[]
while t<100:
    x.append(t)
    yA.append(A)
    yB.append(B)
    yC.append(C)
    dA=-k1*A*dt
    dB=-k2*B*dt+k1*A*dt
    dC=+k2*B*dt
    A+=dA
    B+=dB
    C+=dC
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.plot(x,yC)
plt.xlabel('Tiempo(min)')
plt.ylabel('ConcentaciÃ³n')
plt.title('Cinetica de segundo orden')
plt.show()
```
