```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fraccion(pH,pKa):
    
    H=10**(-pH)
    Ka=[1]
    for pk in pKa:
        x=10**(-pk)
        Ka.append(x)
    n=len(Ka)-1
    mult=1
    terminos=[]
    for a in range(n+1):
        n1=(n-a)
        cH=H**n1
        mult=mult*Ka[a]
        d=cH*mult
        terminos.append(d)
        sumaTerminos=sum(terminos)
    alpha=terminos/sumaTerminos
    return alpha
```

```python
pKa=np.array([4.0,6.7,8.5])
pH=0
x=[]
y0=[]
y1=[]
y2=[]
y3=[]
while pH<=14: #Para calcular en pH de 0.1 en 0.1
    x.append(pH)
    alpha=fraccion(pH,pKa)
    y0.append(alpha[0])
    y1.append(alpha[1])
    y2.append(alpha[2])
    y3.append(alpha[3])
    
    pH+=0.1
plt.plot(x,y0)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlabel('pH')
plt.ylabel('fracciones de las especies')
plt.title('Fracciones de  especies Ã¡cidas')
plt.show()
```
