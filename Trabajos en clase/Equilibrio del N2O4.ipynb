```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deltaS_N2O4=304.276
deltaH_N2O4=9.079*1000
deltaS_NO2=240.034
deltaH_NO2=33.095*1000

T=298.15 #en Kelvin
pt=1.0   #bar
R=8.314  #J/K-mol
```

```python
uN2O4_est= deltaH_N2O4-T*deltaS_N2O4
uNO2_est= deltaH_NO2-T*deltaS_NO2
print(uN2O4_est,uNO2_est)
```

```python
x=np.arange(0.01,1,0.01)
n_N2O4=1-x
n_NO2=2*x
nT=1+x

X_N2O4=n_N2O4/nT
X_NO2=n_NO2/nT

```

# MÃ©todo 1

```python
uN2O4= uN2O4_est+R*T*np.log(X_N2O4)
uNO2= uNO2_est+R*T*np.log(X_NO2)
uT=n_N2O4*uN2O4+n_NO2*uNO2

minimo=np.min(uT)
n=np.where(minimo==uT)
xeq=x[n]
print(xeq)

plt.plot(x,uT)
plt.xlabel(r'$\xi$  (mol)')
plt.ylabel(r'$\Delta G $ (J/mol)')
plt.axvline(x=xeq,color='purple',linestyle='-.')
plt.text(xeq,-78000,r'$\xi_{eq})=$'+str(xeq),fontsize=14.5)
plt.show()
```

```python
pNO2=2*xeq/(1+xeq)*pt
pN2O4=(1-xeq)/(1+xeq)*pt
kp=pNO2**2/pN2O4
print(kp)
```

# MÃ©todo 2

```python
yy=(uN2O4-2*uNO2)**2
minimo=np.min(yy)
n=np.where(yy==minimo)
xeq=x[n]
print(xeq)

plt.plot(x,uN2O4)
plt.plot(x,2*uNO2)
plt.axvline(x=xeq,color='purple',linestyle='-.')
plt.text(xeq,-78000,r'$\xi_{eq})=$'+str(xeq),fontsize=14.5)
plt.xlabel(r'$\xi$  (mol)')
plt.ylabel(r'$\Delta G $ (J/mol)')
plt.show()
```
